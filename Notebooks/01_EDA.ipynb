# **Data Distribution**

## ***Age Distribution***

```python
df["age"].describe()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['age'], bins=20, kde=True)
plt.title('Distribution des âges des joueurs')
plt.xlabel('Âge')
plt.ylabel('Nombre de joueurs')
plt.grid(True)
plt.show()
```

```python
df[df['age']==43][['name','position','nationality','team']]
```

La distribution des âges des joueurs FIFA-21 est centrée autour de 23 à 26 ans, reflétant le pic de performance des footballeurs. Elle diminue après 30 ans, avec peu de joueurs au-delà de 35 ans, majoritairement des gardiens. Quelques jeunes de 17 ans apparaissent comme de futurs talents. Cela illustre bien le cycle de vie typique d’une carrière footballistique.

## **Overall**

```python
plt.figure(figsize=(12,6))
sns.histplot(data=df, x='overall', hue='age', multiple='stack', bins=30, palette='tab20')
plt.title('Distribution de l\'Overall selon l\'âge (histogramme empilé)')
plt.show()


```

```python
# Ou juste KDE (pour les âges les plus fréquents)
ages_to_plot = df['age'].value_counts().head(5).index
plt.figure(figsize=(12,6))
for age in ages_to_plot:
    subset = df[df['age'] == age]
    sns.kdeplot(subset['overall'], label=f'Age {age}')
plt.title('Densité de l\'Overall pour les 5 âges les plus fréquents')
plt.legend()
plt.show()
```

```python
top_teams = df['team'].value_counts().head(10).index
mean_overall_team = df[df['team'].isin(top_teams)].groupby('team')['overall'].mean().sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x=mean_overall_team.index, y=mean_overall_team.values)
plt.title('Moyenne de l\'Overall des 10 équipes les plus représentées')
plt.ylabel('Moyenne Overall')
plt.xlabel('Équipe')
plt.xticks(rotation=45)
plt.show()
```

```python
top_nations = df['nationality'].value_counts().head(10).index
mean_overall_nat = df[df['nationality'].isin(top_nations)].groupby('nationality')['overall'].mean().sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x=mean_overall_nat.index, y=mean_overall_nat.values)
plt.title('Moyenne de l\'Overall des 10 nationalités les plus représentées')
plt.ylabel('Moyenne Overall')
plt.xlabel('Nationalité')
plt.xticks(rotation=45)
plt.show()
```

```python
top5_teams = df.groupby('team')['overall'].mean().sort_values(ascending=False).head(5)
print("Top 5 équipes selon la moyenne de l'overall :")
print(top5_teams)
```

## **Potential**

```python
# --- 1. Distribution du potentiel global (histogramme) ---
plt.figure(figsize=(10,5))
sns.histplot(df['potential'], bins=30, kde=True, color='blue')
plt.title('Distribution globale du potentiel')
plt.xlabel('Potential')
plt.ylabel('Nombre de joueurs')
plt.show()
```

```python
# --- 2. Moyenne du potentiel par âge (barplot) ---
mean_potential_age = df.groupby('age')['potential'].mean().sort_index()
plt.figure(figsize=(12,6))
sns.barplot(x=mean_potential_age.index, y=mean_potential_age.values)
plt.title('Moyenne du potentiel par âge')
plt.xlabel('Âge')
plt.ylabel('Moyenne du potentiel')
plt.show()
```

```python
# --- 3. Moyenne du potentiel par position (barplot) ---
mean_potential_pos = df.groupby('position')['potential'].mean().sort_values(ascending=False)
plt.figure(figsize=(12,6))
sns.barplot(x=mean_potential_pos.index, y=mean_potential_pos.values)
plt.title('Moyenne du potentiel par position')
plt.xlabel('Position')
plt.ylabel('Moyenne du potentiel')
plt.xticks(rotation=45)
plt.show()
```

```python
top_teams = df['team'].value_counts().head(10).index
mean_potential_team = df[df['team'].isin(top_teams)].groupby('team')['potential'].mean().sort_values(ascending=False)
plt.figure(figsize=(12,6))
sns.barplot(x=mean_potential_team.index, y=mean_potential_team.values, palette='coolwarm')
plt.title('Moyenne du potentiel des 10 équipes les plus représentées')
plt.xlabel('Équipe')
plt.ylabel('Moyenne du potentiel')
plt.xticks(rotation=45)
plt.show()
```

```python



# --- 5. Moyenne du potentiel par nationalité (barplot) pour top 10 nationalités ---
top_nations = df['nationality'].value_counts().head(10).index
mean_potential_nat = df[df['nationality'].isin(top_nations)].groupby('nationality')['potential'].mean().sort_values(ascending=False)
plt.figure(figsize=(12,6))
sns.barplot(x=mean_potential_nat.index, y=mean_potential_nat.values, palette='cubehelix')
plt.title('Moyenne du potentiel des 10 nationalités les plus représentées')
plt.xlabel('Nationalité')
plt.ylabel('Moyenne du potentiel')
plt.xticks(rotation=45)
plt.show()
```
